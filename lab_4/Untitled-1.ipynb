```python
from ucimlrepo import fetch_ucirepo, list_available_datasets
import os
from urllib.request import urlopen, Request
from urllib.error import HTTPError
from urllib.parse import urlparse
import zipfile

import pandas as pd
```

```python
class Data:
    DATA_DIRECTORY = "./data"
    DATA_ZIPNAME = ""
    DATA_FILENAME = "household_power_consumption.txt"
    DATA_URL = "https://archive.ics.uci.edu/static/public/235/individual+household+electric+power+consumption.zip"

    def __init__(self, update_data):
        self.update_data = update_data


    def Download(self):
        try:

            if not (os.path.exists(self.DATA_DIRECTORY) or os.path.isdir(self.DATA_DIRECTORY)):
                print(f"Folder for data {self.DATA_DIRECTORY} does not exist. Creating.")
                os.mkdir(self.DATA_DIRECTORY)

            parsed_data_url = urlparse(self.DATA_URL)
            self.DATA_ZIPNAME = os.path.basename(parsed_data_url.path)

            request = Request(self.DATA_URL, method="GET")
            responce = urlopen(request)

            if responce.status != 200:
                raise Exception("Server respond with not 200")
            
            raw_data = responce.read()
            data_filepath = os.path.join(self.DATA_DIRECTORY, self.DATA_ZIPNAME)

            with open(data_filepath, 'wb') as data_file:
                data_file.write(raw_data)

        except HTTPError as e:
            print(f"Error code: {e.code}")
            print(e.read())

        except Exception as e:
            print(f"Error: {e}")
    
    def ExtractZipFile(self):
        try:
            data_filepath = os.path.join(self.DATA_DIRECTORY, self.DATA_ZIPNAME)
            with zipfile.ZipFile(data_filepath, 'r') as data_zip_file:
                data_file = data_zip_file.filelist[0]
                data_zip_file.extract(data_file.filename, self.DATA_DIRECTORY)
                self.DATA_FILENAME = data_file.filename
        
        except zipfile.BadZipFile as e:
            print(e)


    def CreateDataFrame(self):
        data_filepath = os.path.join(self.DATA_DIRECTORY, self.DATA_FILENAME)
        df = pd.read_csv(data_filepath, sep=';')
        return df
            

```
